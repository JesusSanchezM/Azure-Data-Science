# Loadind the data with pandas

```python
import pandas as pd

df_flights = pd.read_csv('https://raw.githubusercontent.com/MicrosoftDocs/ml-basics/master/challenges/data/flights.csv')
df_flights.head()
```

# Reviewing for missing values 

```python
df_flights[['DepDelay', 'ArrDelay']].isnull().sum()

df_flights.isnull().sum()
```

# There's no missing values for DepDelay and ArrDelay. Let's check for outliers

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots(1, 2, figsize = (12,4))

ax[0].boxplot(df_flights['DepDelay'])
ax[1].boxplot(df_flights['ArrDelay'])

print(df_flights[['DepDelay','ArrDelay']].describe())

for col_name in ['DepDelay','ArrDelay']:
    col = df_flights[col_name]
    rng = col.max() - col.min()
    var = col.var()
    std = col.std()
    print('\n{}:\n - Range: {:.2f}\n - Variance: {:.2f}\n - Std.Dev: {:.2f}'.format(col_name, rng, var, std))
```

# Dealing with outliers

```python
import numpy as np

df_1 = df_flights[(df_flights['DepDelay']<=np.mean(df_flights['DepDelay'])+1*np.std(df_flights['DepDelay'])) & 
                  (df_flights['DepDelay']>=np.mean(df_flights['DepDelay'])-1*np.std(df_flights['DepDelay']))].DepDelay

print(df_1)

df_2_D = df_flights[(df_flights['DepDelay']>=-4) & (df_flights['DepDelay']<=9)].DepDelay

print('df_1 count: {}'.format(df_1.count()))
print('df_2 count: {}'.format(df_2_D.count()))
print('Total cases: {}' .format(df_flights['DepDelay'].count()))

fig, ax = plt.subplots(1, 3, figsize = (12,4))

ax[0].boxplot(df_flights.DepDelay)
ax[1].boxplot(df_1)
ax[2].boxplot(df_2_D)

print('-----------------------------------------------------------------------------------------------------------')

df_3 = df_flights[(df_flights['ArrDelay']<=np.mean(df_flights['ArrDelay'])+1*np.std(df_flights['ArrDelay'])) & 
                  (df_flights['ArrDelay']>=np.mean(df_flights['ArrDelay'])-1*np.std(df_flights['ArrDelay']))].DepDelay

print(df_3)

df_4_A = df_flights[(df_flights['ArrDelay']>=-11) & (df_flights['ArrDelay']<=10)].ArrDelay

print('df_3 count: {}'.format(df_3.count()))
print('df_4 count: {}'.format(df_4_A.count()))
print('Total cases: {}' .format(df_flights['ArrDelay'].count()))

fig, ax = plt.subplots(1, 3, figsize = (12,4))

ax[0].boxplot(df_flights.ArrDelay)
ax[1].boxplot(df_3)
ax[2].boxplot(df_4_A)
```

# This is the final data to work with, but I'll modify the data to eliminate all rows that contain missing values and using data from percentile 0.01 to 0.90

```python
df_flights.DepDel15 = df_flights.DepDel15.fillna(0)
df_flights.isnull().sum()
```

```python
# Function to show summary stats and distribution for a column
def show_distribution(var_data):
    from matplotlib import pyplot as plt

    # Get statistics
    min_val = var_data.min()
    max_val = var_data.max()
    mean_val = var_data.mean()
    med_val = var_data.median()
    mod_val = var_data.mode()[0]

    print(var_data.name,'\nMinimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(min_val,
                                                                                            mean_val,
                                                                                            med_val,
                                                                                            mod_val,
                                                                                            max_val))

    # Create a figure for 2 subplots (2 rows, 1 column)
    fig, ax = plt.subplots(2, 1, figsize = (10,4))

    # Plot the histogram   
    ax[0].hist(var_data)
    ax[0].set_ylabel('Frequency')

    # Add lines for the mean, median, and mode
    ax[0].axvline(x=min_val, color = 'gray', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mean_val, color = 'cyan', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=med_val, color = 'red', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mod_val, color = 'yellow', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=max_val, color = 'gray', linestyle='dashed', linewidth = 2)

    # Plot the boxplot   
    ax[1].boxplot(var_data, vert=False)
    ax[1].set_xlabel('Value')

    # Add a title to the Figure
    fig.suptitle(var_data.name)

    # Show the figure
    fig.show()
```

```python
# Trim outliers for ArrDelay based on 1% and 90% percentiles
df_flights = df_flights[df_flights.ArrDelay < df_flights.ArrDelay.quantile(0.90)]
df_flights = df_flights[df_flights.ArrDelay > df_flights.ArrDelay.quantile(0.01)]

# Trim outliers for DepDelay based on 1% and 90% percentiles
df_flights = df_flights[df_flights.DepDelay < df_flights.DepDelay.quantile(0.90)]
df_flights = df_flights[df_flights.DepDelay > df_flights.DepDelay.quantile(0.01)]

# View the revised distributions
delayFields = ['DepDelay','ArrDelay']
for col in delayFields:
    show_distribution(df_flights[col])
```

```python
#1st queastion

print('DepDelay total: {:.2f}' .format(np.mean(df_flights['DepDelay'])))
print('ArrDelay total {:.2f}'.format(np.mean(df_flights['ArrDelay'])))

print('DepDelay filter: {:.2f}'.format(np.mean(df_2_D)))
print('ArrDelay filter: {:.2f}'.format(np.mean(df_4_A)))
```

```python
#2nd queastion

# calculate mean of DepDelay and ArrDelay grouped by Carrier
grouped = df_flights.groupby('Carrier')[['DepDelay', 'ArrDelay']].mean()

# plot the bar chart
grouped.plot(kind='bar', figsize=(12,4))

plt.title('DepDelay and ArrDelay by Carrier')
plt.show()

```

```python
#3erd queastion 

grouped_2 = df_flights.groupby('DayOfWeek')[['ArrDelay']].mean()

grouped_2.plot(kind='bar', figsize=(12,4))

plt.title('ArrDelay per day of week')

df_flights.boxplot(column='ArrDelay', by='DayOfWeek', figsize=(12,4))

plt.show()
```

```python
#4rth queastion

grouped = df_flights.groupby('Carrier')[['DepDelay', 'ArrDelay']].mean()
print(grouped.sort_values('DepDelay', ascending=False))

max_index = grouped['DepDelay'].idxmax()
max_value = grouped.loc[max_index, 'DepDelay']
print(f"Carrier: {max_index}, DepDelay's mean: {max_value}")


```

```python
#6 queastion

grouped_4 = df_flights.groupby(['OriginAirportName', 'DestAirportName'])['ArrDelay'].sum()
group_route = pd.DataFrame(grouped_4.reset_index())

print(group_route.sort_values('ArrDelay', ascending=False))

maxid = group_route['ArrDelay'].idxmax()
origin_airport = group_route.loc[maxid, 'OriginAirportName']
dest_airport = group_route.loc[maxid, 'DestAirportName']
delay = group_route.loc[maxid, 'ArrDelay']

print("Max delay: {} from {} to {}".format(delay, origin_airport, dest_airport))

group_route.sort_values('ArrDelay', ascending=False)
```
