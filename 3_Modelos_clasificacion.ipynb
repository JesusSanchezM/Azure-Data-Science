# Reading dataframe

```python
import pandas as pd

df_n = pd.read_csv('https://raw.githubusercontent.com/MicrosoftDocs/ml-basics/master/challenges/data/wine.csv')
```

```python
df = df_n
```

```python
df_n.head()
```

```python
df.isnull().sum()
```

# No missing values

```python
from matplotlib import pyplot as plt

for x in df:

    feature = df[x]
    print(feature)
    fig = plt.figure(figsize=(12, 6))
    plt.boxplot(feature)
    
fig.show();

```
